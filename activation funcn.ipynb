```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    data = np.tanh(x)
    return data

x = np.linspace(-10,10,100)
y= tanh(x)

plt.plot(x, y)
plt.title('tanh activation function')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    data = 1/(1 + np.exp(-x))
    return data

def swish(x):
    data = x * sigmoid(x)
    return data

x = np.linspace(-10,10,100)
y= swish(x)

plt.plot(x, y)
plt.title('swish activation function')
plt.grid()
plt.show()
```
